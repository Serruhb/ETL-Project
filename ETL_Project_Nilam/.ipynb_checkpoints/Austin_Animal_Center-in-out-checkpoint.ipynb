```python
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
import numpy as np
from config import userid,pwd
```

## Extraction

* creating a file path
* reading file in pandas

```python
#Extract

intake_file = "austin-animal-center-intakes.csv"
outcome_file = "austin-animal-center-outcomes.csv"

intake_df = pd.read_csv(intake_file)
outcome_df = pd.read_csv(outcome_file)

```

```python
#preview outcomes
outcome_df.head()
```

```python
#preview intake
intake_df.head()
```

## Transforamtion for Outcome
* reduce to pertinent columns
* reformat 'DateTime' to reflect needed information
* rename columns for sql

```python
#Viewing columns needed
outcome_cleaned_df = outcome_df[["Animal ID","DateTime","Outcome Type","Animal Type","Age upon Outcome","Breed","Color"]]

outcome_cleaned_df.head()
```

```python
#reformat date (stripped of the timestamp)
outcome_cleaned_df = outcome_cleaned_df.copy()
outcome_cleaned_df["DateTime"] = pd.to_datetime(outcome_cleaned_df["DateTime"]).dt.date
outcome_cleaned_df.head()
```

```python
outcome_table = outcome_cleaned_df
outcome_table = outcome_table.rename(columns = {"Animal ID":"animal_id",
                                                 "DateTime":"datetime",
                                                  "Outcome Type":"outcome_type",
                                                  "Animal Type":"animal_type",
                                                  "Age upon Outcome" : "age",
                                                   "Breed":"breed",
                                                    "Color":"color"})
                                                    
# Clean the data by dropping duplicates and setting the index
outcome_table.drop_duplicates("animal_id", inplace=True)            
# Set index
outcome_table.set_index("animal_id", inplace=True)
outcome_table.head()
```

# Transforamtion for Intakes



```python
#Viewing columns needed
intake_cleaned_df = intake_df[["Animal ID","DateTime","Intake Type","Animal Type","Age upon Intake","Breed","Color"]]

intake_cleaned_df.head()
```

```python
#reformat date (stripped of the timestamp)
intake_cleaned_df = intake_cleaned_df.copy()
intake_cleaned_df["DateTime"] = pd.to_datetime(intake_cleaned_df["DateTime"]).dt.date
intake_cleaned_df.head()
```

```python
intake_table = intake_cleaned_df
intake_table = intake_table.rename(columns = {"Animal ID":"animal_id",
                                                 "DateTime":"datetime",
                                                  "Intake Type":"intake_type",
                                                   "Animal Type":"animal_type",
                                                   "Age upon Intake" : "age",
                                                   "Breed":"breed",
                                                    "Color":"color"})
# Clean the data by dropping duplicates and setting the index
intake_table.drop_duplicates("animal_id", inplace=True)            
# Set index
intake_table.set_index("animal_id", inplace=True)
intake_table.head()
intake_table.head()
```

```python
#FOR MERGE
#list of common columns

common_columns = ["animal_id","datetime","animal_type","age","breed","color"]
intake_outcomes_merged = pd.merge(intake_table,outcome_table,on=common_columns)
intake_outcomes_merged.head()
```
